# Object Oriented Programming

```python
class coordinate(object):
    # the __init__ function is a constructor that you call on creation
    def __init__(self, x, y): # self is equivalent to this (own data)
        self.x = x
        self.y = y
    
    # methods
    def grid(self): # self param is always needed
        gridwid = "#"
        bot = "0#"
        num = " 0"
        for i in range(self.y):
            gridwid += ' '
            bot += '#'
            num += str(i+1)
        print(str(self.y)+gridwid[:-1]+'X')
        for i in range(self.y-1):
            print(str(self.y-i-1)+gridwid)
        print(bot)
        print(num)
    
    
    


location = coordinate(5, 5) # creating an instance of a class - two parameters are passed into _init_ function (self is skipped)
print(location.x) # self.x of location

location.grid() # call the method

# call methods without an instance:
coordinate.grid(location) # pass in a coordinate instance instead of nothing to replace 'self'
```

```python
class example(object):
    # builtin functions
    def __init__(self, a, b): # init is called when creating a new class instance
        self.a = a
        self.b = b
        print('example class initiated')
    
    def __str__(self): # used when printing a class instance
        return '<' + str(self.a) + ', ' + str(self.b) + '>'

anexample = example(100, 299)

print(anexample)
```

```python
print(isinstance(anexample, example)) # test if is anexample is an instance of example
print(type(anexample)) # print the type of anexample
```

```python
class biggerexample(object):
    def __init__(self, a, b):
        self.a = a
        self.b = b
    def __str__(self): # print(self)
        return f'<{self.a}, {self.b}>'
    def __add__(self, other): # self + other
        return self.a+other.a*(self.a+other.b)
    def __sub__(self, other): # self - other
        return self.a-other.a*(self.a-other.b)
    def __eq__(self, other): # self == other
        return self.a == other.a and self.b == other.b
    def __lt__(self, other): # self < other
        return self.a < other.a
    def __len__(self): # len(self)
        return self.a

    # many more
    


alargerexample = biggerexample(100, 99)
print(alargerexample)
print(len(alargerexample))

exampletwo = biggerexample(66, 22)
print(f'''
{alargerexample+exampletwo}
{alargerexample-exampletwo}
{alargerexample==exampletwo}
{alargerexample<exampletwo}
''')

```

```python
class Fraction(object):
    def reducefract(n, d):
        def gcd(n, d):
            while d != 0:
                t = d
                d = n%d
                n = t
            return n
        greatest=gcd(n,d)
        n/=greatest
        d/=greatest
        return Fraction(int(n), int(d))

    def __init__(self, num, den):
        self.n = num
        self.d = den
    
    def __str__(self):
        return f'{self.n}/{self.d}'
    
    def __int__(self):
        return round(self.n/self.d)
    def __float__(self):
        return self.n/self.d
    
    def __add__(self, other):
        newn = self.n*other.d+other.n*self.d
        newd = self.d*other.d
        return Fraction.reducefract(newn, newd)
    def __mul__(self, other):
        newn = self.n*other.n
        newd = self.d*other.d
        return Fraction.reducefract(newn, newd)
    def __sub__(self, other):
        newn = self.n*other.d-other.n*self.d
        newd = self.d*other.d
        return Fraction.reducefract(newn, newd)
    def __truediv__(self, other):
        newn = self.n*other.d
        newd = self.d*other.n
        return Fraction.reducefract(newn, newd)

alpha = Fraction(3, 4)
beta = Fraction(6, 4)

print(f'''
{alpha+beta}
{alpha*beta}
{alpha-beta}
{alpha/beta}
{int(alpha)}
{float(beta)}
''')
```
